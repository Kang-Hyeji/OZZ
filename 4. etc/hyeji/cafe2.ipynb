```
from konlpy.tag import Okt
okt = Okt()
```

```
cafe = {
    "에스프레소" : 3000,
    "아메리카노": 4000,
    "아아": 4000,
    "카페라떼" : 4500,
    "바닐라라떼":5000,
    "아바라":5000,
    "레몬에이드" : 5500,
    "자몽차":5000,
    "초코라떼" : 5000,
    "자바칩프라프치노":5500,
    "녹차라떼":5000,
    "치즈케익":4000,
    "카푸치노" : 4500,
    "밀크티" : 4500,
    "베이글" : 3000,
    "샌드위치" : 3500
}


number = {
    "한잔":1,
    "한개":1,
    "두잔":2,
    "두개":2,
    "세잔":3,
    "세개":3,
    "네잔":4,
    "네개":4,
    "다섯잔":5,
    "다섯개":5
}
```

```
text = input()

text_token = okt.nouns(text)

def remove_values_from_list(the_list, val):
   return [value for value in the_list if value != val]

text_token = remove_values_from_list(text_token, '아이스')
sum = 0
check=''
# print(okt.pos(text))
# print(okt.nouns(text))
len_order = len(text_token)


for i in range(0,len_order,2) :
  if text_token[i] in cafe:
    menu= text_token[i]
    num = number[text_token[i+1]]
    sum += cafe[menu]*num
    check += menu+' '+text_token[i+1]+','
  else :
    print("죄송합니다. 저희 메뉴에 ",text_token[i],"가 없습니다 ㅠㅠ")

print(check,"총 가격은 ",sum,"원 입니다")
  
```
